```python
import numpy as np
```

# To add, subtract, multiply, divide arguments element-wise:

```python
array1=np.random.randint(1,10,3)
array2=np.array([4,5,6])
```

```python
add= np.add(array1,array2)
print(add)
```

```python
sub=np.subtract(array1,array2)
print(sub)
```

```python
mul=np.multiply(array1,array2)
print(mul)
```

```python
div=np.divide(array1,array2)
print(div)
```

# To round elements of the array to the nearest integer:

```python
y = np.array([0.2, 2.3, 0.4, 0.5, 2.6, 0.7])
y = np.rint(y)
print(y)
```

# To calculate mean across dimension, in a 2D numpy array:

```python
import numpy as np

a = np.array([np.random.randint(1, 5, 3), np.random.randint(1, 5, 3)])

print(a)
```

```python
b=np.mean(a,axis=0) 
c=np.mean(a,axis=1) 
print(b)
print(c)
```

# To calculate the difference between neighboring elements, element-wise of a given array:

```python
# create a numpy 1d-array
arr = np.array([1, 12, 3, 14, 5,16, 7, 18, 9, 110])
```

```python
# finding the difference between neighboring elements
result = np.diff(arr)
print(result)
```
