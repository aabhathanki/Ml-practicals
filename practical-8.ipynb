```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Write a Pandas program to create a line plot of the opening, closing stock prices of given company between two specific dates:

```python
# Generate random opening and closing prices for a year
data =pd.read_csv('/kaggle/input/stock-market-data-nifty-50-2000-2022/Nifty_50_with_indicators_.csv')
# Plot opening and closing prices
data.plot(x='Date', y=['open', 'close'])
# Set labels and show the plot
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```
